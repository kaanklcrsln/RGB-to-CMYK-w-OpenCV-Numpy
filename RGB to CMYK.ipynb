```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)

h, w = gray_image.shape
pixels = gray_image.flatten()

plt.figure(figsize=(10, 6))
plt.hist(pixels, bins=256, color="gray", alpha=0.7, edgecolor="black")
plt.title("Grayscale Yoğunluk Grafiği")
plt.xlabel("Yoğunluk Değeri (0-255)")
plt.ylabel("Piksel Sayısı")
plt.grid(alpha=0.3)
plt.show()

x = np.arange(len(pixels)).reshape(-1, 1)  
y = pixels.reshape(-1, 1)  

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

plt.figure(figsize=(14, 6))
plt.scatter(x, y, s=1, color="blue", alpha=0.5, label="Gerçek Piksel Değerleri")
plt.plot(x, y_pred, color="red", linewidth=2, label="Regresyon Çizgisi")
plt.title("Piksel Yoğunluk Analizi ve Regresyon Grafiği")
plt.xlabel("Piksel İndeksi")
plt.ylabel("Yoğunluk Değeri (0-255)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Regresyon Eğimi (Slope): {model.coef_[0][0]:.4f}")
print(f"Regresyon Kesim Noktası (Intercept): {model.intercept_[0]:.4f}")
```
